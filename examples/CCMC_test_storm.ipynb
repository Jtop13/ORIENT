```python
import ORIENT as orient

from datetime import datetime
import warnings

# Suppress pandas fragmentation warnings
warnings.filterwarnings("ignore", message=".*DataFrame is highly fragmented.*")

start_time = datetime(2013,5,15)
end_time = datetime(2013,6,15)
input_time =  datetime(2013,6,1,3)

eflux_1 =  orient.eflux.model.ElectronFlux(start_time, end_time,instrument = 'mageis',channel = 11)
final_frame, X_input_total = eflux_1.get_flux(dst_source='omni',
                 al_source='omni',
                 sw_source='omni',use_omni = True,use_traj = False,get_input_time = input_time,get_MLT_flux=True,
                                             selected_MLT_datetime = input_time)
```

```python
eflux_2 =  orient.eflux.model.ElectronFlux(start_time, end_time,instrument = 'mageis',channel = 14)
final_frame2, X_input_total2 = eflux_2.get_flux(dst_source='omni',
                 al_source='omni',
                 sw_source='omni',use_omni = True,use_traj = False,get_input_time = input_time,get_MLT_flux=True,
                                             selected_MLT_datetime = input_time)

eflux_3 =  orient.eflux.model.ElectronFlux(start_time, end_time,instrument = 'mageis',channel = 16)
final_frame3, X_input_total3 = eflux_3.get_flux(dst_source='omni',
                 al_source='omni',
                 sw_source='omni',use_omni = True,use_traj = False,get_input_time = input_time,get_MLT_flux=True,
                                             selected_MLT_datetime = input_time)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1.inset_locator import inset_axes

# Set global font size and line width
plt.rcParams['lines.linewidth'] = 2
plt.rcParams['axes.labelsize'] = 22         # Axis label font size
plt.rcParams['axes.titlesize'] = 24         # Axis title font size
plt.rcParams['xtick.labelsize'] = 20        # X tick label font size
plt.rcParams['ytick.labelsize'] = 20        # Y tick label font size
plt.rcParams['legend.fontsize'] = 20        # Legend font size
plt.rcParams['figure.titlesize'] = 26       # Figure title font size

fig, axs = plt.subplots(8, 1, figsize=(20, 16), sharex=True)
plt.subplots_adjust(hspace=0)

# Line plots with axis labels
axs[0].plot(eflux_1.dst.index, eflux_1.dst.values, color='black')
axs[0].set_ylabel('SYMH')

axs[1].plot(eflux_1.flow_speed.index, eflux_1.flow_speed.values, color='black')
axs[1].set_ylabel('$V_{SW}$')

axs[2].plot(eflux_1.pressure.index, eflux_1.pressure.values, color='black')
axs[2].set_ylabel('$P_{SW}$')

axs[3].plot(eflux_1.bz.index, eflux_1.bz.values, color='black')
axs[3].set_ylabel('$B_{z}$')

axs[4].plot(eflux_1.al.index, eflux_1.al.values, color='black')
axs[4].set_ylabel('Modeled AL')

axs[5].set_ylabel('L shell')
axs[6].set_ylabel('L shell')
axs[7].set_ylabel('L shell')

# Flux plots
ob1 = eflux_1.make_flux_plot(axs[5], normmax=10**4.5, normmin=1)
ob2 = eflux_2.make_flux_plot(axs[6], normmax=10**4.5, normmin=1)
ob3 = eflux_3.make_flux_plot(axs[7], normmax=10**4.5, normmin=1)

# Add colorbar
color_axis = inset_axes(axs[7], width="1%", height="300%", loc='lower left',
                         bbox_to_anchor=(1.01, 0., 1, 1), bbox_transform=axs[7].transAxes,
                         borderpad=0)
fig.colorbar(ob1, cax=color_axis, label=r'$cm^{-2}s^{-1}sr^{-1}keV^{-1}$')

# Energy level labels (last 3 subplots)
energy_levels = ["235 keV", "597 keV", "909 keV"]
for i, energy in enumerate(energy_levels):
    axs[i+5].text(0.01, 0.9, energy, transform=axs[i+5].transAxes, fontsize=20, verticalalignment='top', color='black')

# Source labels (first 5 subplots)
from_source = ["OMNI", "OMNI", "OMNI", "OMNI", "OMNI"]
for i, source in enumerate(from_source):
    axs[i].text(1.01, 0.8, "From", transform=axs[i].transAxes, fontsize=18, verticalalignment='top', color='black')
    axs[i].text(1.01, 0.6, source, transform=axs[i].transAxes, fontsize=18, verticalalignment='top', color='black')

# Set x-axis limit
axs[0].set_xlim(eflux_1.t_data[0], eflux_1.t_data[-1])

# Save figure
plt.savefig('CCMC_test_2013.jpg', format='jpeg', bbox_inches='tight')
```
